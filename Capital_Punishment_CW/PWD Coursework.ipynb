## Step 1: Preparing the first dataset

### Explanation as to why this data source is appropriate for the research question posed.

This data source contains all the data for prison population for various countries over the years of 2003 to 2018. The reason why I have picked this dataset is as follows:

Firstly, its coluumn heading includes total persons held in prisons by country, counts and rates per 100,000 population. Hence, it is relevant to my research question as I can use prison population in rates as a means to indicate crime rates of a country. This is because almost all crimes, from minor to major offenses, including those that include death row, require criminals to be sent to prison. Hence, I find that it is a good indication to the number of crimes that have been committed in a country, per 100,000 population.

Secondly, I have chosen to find a data source that excludes the last two years of 2019-2020, which are years affected by the pandemic. This is because, depending on how a country handles the pandemic, crime rates have increased differently for such countries, hence skewing the dataset to various different degrees. 

### Origin of first data source

This data originates from the United Nations Office on Drugs and Crime (UNODC) and is sourced from the website with url = https://dataunodc.un.org/data/prison/persons%20held%20total. The acquisition technique used is by downloading an excel sheet and changing it into a dataframe for further analysis later. This was the next best alternative as downloading it as a csv made it unable to be acquired into a data frame. 

Since this data can be easily found on the web, it is from an open source. The author is anonymous.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import library to plot graphs later
```

```python
crime_rates_df = pd.read_excel("prison population 2003-2018.xls")
# reading exel sheet

crime_rates_df.head()
# print dataframe with excel sheet data
```

### Cleaning data frame to only include prison rates from the decade 2009 to 2018
 
Since there are many political and economical changes over the years which are main attributes that cause changes in crime rates, I have decided to only include data from the last decade and not before that. Also, the reason why I chose a span of ten years is because I am trying to find a more accurate and recent crime indicator for each country, however there is missing data over the years. Hence, I am trying to make the average value more accurate by using data from more years. 

In the following steps, I will show how I will clean the data source to only include data from the decade 2009 to 2018.

```python
columns = ['Unnamed: 2','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1',]
crime_rates_df=crime_rates_df[columns] 
# extracting data from the years 2009-2018

crime_rates_df.head()
```

```python
crime_rates_df=crime_rates_df.drop(index = crime_rates_df.index[0], axis=0)
# dropping the first row

crime_rates_df = crime_rates_df.rename(columns={"Unnamed: 2": "Country", "2009.1": "2009", "2010.1": "2010", 
                        "2011.1": "2011", "2012.1": "2012", "2013.1": "2013", "2014.1": "2014", 
                        "2015.1": "2015", "2016.1": "2016", "2017.1": "2017", "2018.1": "2018"})
# rename columns

crime_rates_df.head()
```

```python
mean_crimeRate = crime_rates_df.copy()
# copy to a new table to preserve original crime rate table for use later

mean_crimeRate['Mean'] = mean_crimeRate.iloc[:, 1:-1].mean(axis=1)
# finding mean value of crime rates for each country

mean_crime_rates_df = mean_crimeRate.sort_values("Mean", ascending=False)
# sort by descending for mean values

mean_crime_rates_df.index = range(225)
# arrange index in ascending order

mean_crime_rates_df.head()
```

### Cleaning data frame to show country and mean

```python
columns = ['Country','Mean',]
mean_crime_rates_df = mean_crime_rates_df[columns] 
# cleaning data frame to show country and mean

mean_crime_rates_df = mean_crime_rates_df.dropna()
# removing data that has null as its mean value

mean_crime_rates_df.head()
```

## Step 2: Preparing the second dataset

### Explanation as to why this data source is appropriate for the research question posed.

This data source is chosen as the column header includes the status of a country on whether it adopts or have abolished the death penalty in the year 2020. Hence, the information is current and serves as an accurate indicator as to whether the death penalty is still in place crime rates were recorded in the last decade, and is thus useful to the research.

### Origin of second data source

This data originates from WorldData.info and is sourced from the website with url = https://www.worlddata.info/deathpenalty.php. The acquisition technique used is by webscraping data from a php website and changing it into a dataframe for further analysis.

Since this data can be easily found on the web, it is from an open source. The author is anonymous.

```python
import requests

url="https://www.worlddata.info/deathpenalty.php"
response = requests.get(url)

response.text
```

```python
new_table = pd.read_html(response.text)[0]

new_table.head()
# web scraping with php
```

```python
new_table.to_csv("Countries and death sentences")
death_sentences_df = pd.read_csv("Countries and death sentences")
# create new file to save dataset as csv

death_sentences_df.head()
```

### Cleaning dataframe to only show country name, and whether or not the country still adopts capital punishment

```python
columns = ['Country','Status']
death_sentences_df = death_sentences_df[columns] 
# cleaning dataframe to only include country and status column

death_sentences_df.head()
```

## Step 3: Analysis 1 - Analysing the average crime rate for each country

Divide data into two datasets - one with high crime rates (HCR) and another with low crime rates (LCR) as an average for each country from 2009-2018. Analyse the data using a bar chart to compare the number of countries with and without death penalty that have higher or lower crime rates.

### Merging table to show correlation between both datasets

```python
correlation_table = pd.merge(mean_crime_rates_df, death_sentences_df, how='inner', on='Country')
# merging both datasets into one new table by using name of country as ID tag

correlation_table.head()
```

### Preparing dataframe to be used directly for analysis 

Finding the median value for average crime rates to be used as a basis to split table into two sections for analysis 

```python
median = correlation_table["Mean"].median()
median
```


# Find median value

Splitting data into two tables. Data with crime rates above the median value will be labelled as countries with comparatively higher crime rates (HCR) and data with crime rates below the median value will be labelled as countries with comparatively lower crime rates (LCR).

Since there is a uneven number of rows, the median value is removed from the dataframe and excluded from both tables to ensure equal number of data for even comparison.

```python
hcr_df = correlation_table[correlation_table['Mean'] > median]
# creating a dataframe with the highest half crime rates

hcr_df
```

```python
hcr_active = hcr_df[hcr_df['Status'].isin(['active','only for serious crimes'])]
# create dataframe which includes data for active capital punishment for countries with high crime rates

hcr_active.index = range(28)

hcr_active.head()
```

```python
hcr_abolished = hcr_df[hcr_df['Status'].isin(['abolished in practice'])]
# create dataframe which includes data for abolished capital punishment for countries with high crime rates

hcr_abolished.index = range(8)

hcr_abolished.head()
```

```python
lcr_df = correlation_table[correlation_table['Mean'] < median]
# creating a dataframe with the lowest half crime rates

lcr_df.index = range(36)

lcr_df.head()
```

```python
lcr_active = lcr_df[lcr_df['Status'].isin(['active','only for serious crimes'])]
# create dataframe which includes data for active capital punishment for countries with low crime rates

lcr_active.index = range(24)

lcr_active.head()
```

```python
lcr_abolished = lcr_df[lcr_df['Status'].isin(['abolished in practice'])]
# create dataframe which includes data for abolished capital punishment for countries with low crime rates

lcr_abolished.index = range(12)

lcr_abolished.head()
```

### Creating a bar graph for analysis

```python
hcr_active_count = len(hcr_active) 
hcr_abolished_count = len(hcr_abolished) 
lcr_active_count = len(lcr_active) 
lcr_abolished_count = len(lcr_abolished)
# creating a python dictionary

high_crime_rate = [hcr_active_count, hcr_abolished_count]
low_crime_rate = [lcr_active_count, lcr_abolished_count]

index = ['Active capital punishment', 'Abolished capital punishment']

correlation_bargraph = pd.DataFrame({'High crime rate':high_crime_rate,'Low crime rate':low_crime_rate }, index = index )
# creating a bar graph

ax = correlation_bargraph.plot.bar(rot=0)
```

## Step 4: Conclusion based on first analysis

From the first analysis, we have divided the dataset into two dataframes - one with higher crime rates and another with lower crime rates, evenly, with 36 sets of data each, for fair comparison. After plotting a bar graph, we can hence say that for countries with active capital punishment, more of them fall into the dataset with higher crime rates (28 countries vs 24) , and for the countries with abolished capital punishment, there are more countries that fall into the dataset with lower crime rates (12 countries vs 8). 

Hence, we can draw the conclusions that active capital punishment results in more countries with higher crime rates and abolishing the capital punishment results in more countries with lower crime rates. Hence, this not only agrees with my hypothesis that there is no correlation between implementing capital punishment and lower crime rates, it also goes beyond to suggest that capital punishment is indeed ineffective in reducing crime.

### Problems/biasness within the dataset
One possible flaw with the technique is that after cleaning and merging both datasets, the number of data available became  reduced such that sample size becomes smaller. Hence, one might argure that it is not accurate enough to be representative of all countries of the world to draw a meaningful enough conclusion.

## Step 5: Analysis 2 - Using the average crime rate for each year

### Merge both data sources into one table

```python
new_correlation_table = pd.merge(crime_rates_df, death_sentences_df, how='inner', on='Country')
# merging both datasets into one new table by using name of country as ID tag

new_correlation_table.head()
```

```python
active_mcr_by_year = new_correlation_table[new_correlation_table['Status'].isin(['active','only for serious crimes'])]
# create new dataframe with countries that have active capital punishment

active_mcr_by_year.index = range(53)

active_mcr_by_year.head()
```

```python
active_mcr_by_year = new_correlation_table[new_correlation_table['Status'].isin(['active','only for serious crimes'])]
# create new dataframe with countries that have active capital punishment

active_mcr_by_year.index = range(53)

active_mcr_by_year.head()
```

### Clean dataframe to only include crime rates over the years for all countries with active death penalty

```python
active_mcr_by_year_df = active_mcr_by_year.transpose()
# switching row and column to easily find average of the new rows 

active_mcr_by_year_df = active_mcr_by_year_df.drop(index = active_mcr_by_year_df.index[0], axis=1)
# dropping the first row

active_mcr_by_year_df = active_mcr_by_year_df.drop(index = active_mcr_by_year_df.index[-1], axis=1)
# dropping the last row

active_mcr_by_year_df.head()
```

### Find average crime rate of all countries with active capital punishment for each year from 2009 to 2018 

```python
active_mcr_by_year_df['Mean'] = active_mcr_by_year_df.iloc[:, 1:-1].mean(axis=1)
# finding mean value of crime rates for each year

active_mcr_by_year_df.head()
```

### Split dataframe to get data of countries with abolished death penalty

```python
abolished_mcr_by_year = new_correlation_table[new_correlation_table['Status'].isin(['abolished in practice'])]
# create new dataframe with countries that have abolished capital punishment

abolished_mcr_by_year.index = range(21)

abolished_mcr_by_year.head()
```

### Clean dataframe to only include crime rates over the years for all countries with abolished death penalty

```python
abolished_mcr_by_year_df = abolished_mcr_by_year.transpose()
# switching row and column to easily find average of the new rows 

abolished_mcr_by_year_df = abolished_mcr_by_year_df.drop(index = abolished_mcr_by_year_df.index[0], axis=1)
# dropping the first row

abolished_mcr_by_year_df = abolished_mcr_by_year_df.drop(index = abolished_mcr_by_year_df.index[-1], axis=1)
# dropping the last row

abolished_mcr_by_year_df.head()
```

### Find average crime rate of all countries with abolished capital punishment for each year from 2009 to 2018 

```python
abolished_mcr_by_year_df['Mean'] = abolished_mcr_by_year_df.iloc[:, 1:-1].mean(axis=1)
# finding mean value of crime rates for each year

abolished_mcr_by_year_df.head()
```

### Plotting a line graph 

It shows the average crime rates over the years for active and abolish capital punishment respectively.

```python
plt.plot(active_mcr_by_year_df["Mean"], label="Active death penalty")
plt.plot(abolished_mcr_by_year_df["Mean"], label="Abolished death penalty")
# plot individual lines

plt.legend()
plt.xlabel("Years")
plt.ylabel("Crime rates")
# creating a legend

plt.show()
# display plot
```

## Step 6: Conclusion based on second analysis

From the second analysis, we have divided the dataset into two dataframes - one with active death penalties in place and another with abolished death penalties as an average for each year, to observe any trends over the year. Unfortunately, from the line graph above, it is difficult to discern any sort of overall increase or decrease in the trend over the years. For the active death penalty, crime rates were dropping from the years 2009 to 2015 before increasing to its original level in 2018. FOr countries with abolished death penalty, crime rates show a slight increase across the years to 2017, before dropping in 2018. Since there is no discernable trend, this still agrees with my hypothesis that there is no correlation between implementing capital punishment and lower crime rates, as crime rates remain to be increasing and decreasing at a similar rate constantly.

However, even though there is no clear conclusion on the impact the death penalty has on the trend on crime rates across the years, it can be clearly seen that in the decade, countries with active death penalty has consistently higher crime rates compared to countries with the abolished death penalty.

Hence, we can draw the conclusions that active capital punishment does indeed results in more countries with higher crime rates and abolishing the capital punishment results in more countries with lower crime rates. Hence, this not only agrees with my hypothesis that there is no correlation between implementing capital punishment and lower crime rates, it also goes beyond to suggest that capital punishment is indeed ineffective in reducing crime.

### Problems/biasness within the dataset
One of the key features of the dataset that might have been problematic is that data is skewed towards the subset of data for countries with active capital punishment. This is because after cleaning the data and splitting the dataset into two new data frames - one for active capital punishment and another for abolished capital punishment, I realised that there are 53 sets of data for the first and 21 for the latter. Hence, there might be an over-representation of data from the former such that when finding average crime rates for both data frames, the latter becomes much less accurate in comparison to the first.
